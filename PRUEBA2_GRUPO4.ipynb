#**EJERCICIO 2**

Importe el dataset athletes del repositorio del curso y realice las siguientes tareas:

Importarlo como nombre athletes
Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
Calcule la masa muscular por fila con una función como
Im.m.=WeightHeight2

Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

¿Cuál es el pais con más participación femenina?

**SOLUCIÓN**

```python
#Importar la libreria requerida
import pandas as pd

#Punto 1:
#Importa el dataset 'athletes'

athletes = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv")
athletes.head()
```

```python
#Punto 2:
#Filtra el datafreme 'athletes' para que solo queden los atletas que han ganado al menos 1 medalla gold, silver o bronze

athletes = athletes.loc[(athletes['gold'] >= 1) | (athletes['silver'] >= 1)| (athletes['bronze'] >= 1)]
athletes.head()
```

```python
#Punto 3:
#Función para calcular la masa muscular

def calcular_masa_muscular(row):
  return row.loc['weight'] / (row.loc['height']**2)

#Aplica la función a cada fila del dataframe
athletes['masa_muscular'] = athletes.apply(calcular_masa_muscular,
         axis = 1)

#Muestra los resultados
athletes.head()
```

```python
#Punto 4:
#Agrupa los datos por la columna 'sport' y calcula la media para el resto de características numericas
stats = athletes[['height','weight','gold','silver','bronze','masa_muscular','sport']].groupby('sport').mean().reset_index()

#Deporte en el que los jugadores tienen mayor peso promedio
id_max = stats.weight.idxmax()
stats.loc[id_max, 'sport']
```

```python
#Punto 5:
#Agrupa los datos por nacionalidad y sexo y arroja el país con más cantidad de mujeres

athletes.groupby(['nationality', 'sex']).sex.count().sort_values(ascending = False).index[0][0]
```

**EJERCICIO 3 PUNTO 1**

Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un DataFrame con el nombre del equipo y la cantidad de dinero de la nómina)

```python
sa = df.groupby(['Team'])[['Salary']].sum().sort_values(by = 'Salary', ascending=False)
```

**EJERCICIO 3 PUNTO 2**

Separe la variable Age en tres grupos y cree una nueva columna llamada Age-Group con esta información.

```python
df['Age Group'] = pd.cut(df['Age'], bins = 3, labels = ['Juniors', 'Pro', 'Master'])
df.head()
```
